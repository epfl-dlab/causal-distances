```python
from collections import defaultdict
import pandas as pd
import seaborn as sns
import numpy as np
import json
import copy
import os
from matplotlib import pyplot as plt
```

```python
palette_4 = ['#601A4A', '#EE442F', '#63ACBE', '#CCBE9F']
sns.palplot(palette_4)
```

```python
with open('baseline_evaluation/results/results-n5.json', 'r') as f:
    results_n5 = json.loads(f.read())
```

```python
with open('baseline_evaluation/results/results-n10.json', 'r') as f:
    results_n10 = json.loads(f.read())
```

```python
real_results = []
for f in os.listdir('real_networks_evaluation/results'):
    with open(os.path.join('real_networks_evaluation/results', f), 'r') as f:
        real_results.extend(json.loads(f.read()))
```

```python
results = copy.copy(results_n5)
results.extend(copy.copy(results_n10))
```

```python
full_results = copy.copy(results)
full_results.extend(real_results)
```

```python
def average_performance(results, metric):
    x, y = [], []
    for outcome in results:
        x.append(outcome['algo'])
        y.append(float(outcome[metric]))
            
    return pd.DataFrame.from_dict({'Algorithm': x, metric: y})
```

```python
def average_performance_samples(results, metric):
    x, y, samples = [], [], []
    for outcome in results:

        if metric == 'ID':
            x.append(outcome['algo'])
            y.append(np.mean(outcome[metric]))
            samples.append(int(outcome['nb_samples']))
        else:
            x.append(outcome['algo'])
            y.append(outcome[metric])
            samples.append(int(outcome['nb_samples']))
            
    return pd.DataFrame.from_dict({'Algorithm': x, metric: y, 'samples': samples})
```

```python
eval_ID_samples = average_performance_samples(results, 'ID')
```

```python
eval_ID_samples_1 = eval_ID_samples[eval_ID_samples['Algorithm'].isin(['LiNGAM', 'CAM', 'CCDr', 'GS'])]
```

```python
eval_ID_samples_2 = eval_ID_samples[eval_ID_samples['Algorithm'].isin(['GES', 'PC', 'IAMB', 'MMPC'])]
```

```python
sns.set_context('paper')
plt.figure(figsize=(10,1))
fig = sns.relplot(x='samples', y='ID', hue='Algorithm', kind="line", ci=35, n_boot=100, palette=palette_4, 
                  data=eval_ID_samples_1, height=5.27, aspect=7.7/5.27,
                 marker='o')
fig._legend.remove()
plt.xlabel('Number of samples', fontsize=17)
plt.ylabel('ID', fontsize=17)
handles, labels = fig.ax.get_legend_handles_labels()
plt.legend(handles=handles[1:], labels=labels[1:], loc='upper center', ncol=2, fontsize=14)
fig.savefig('results/evaluation_samples_1.pdf', dpi=300)
#sns.boxplot(x="Algorithm", y="ID", data=eval_ID_200)
```

```python
def average_performance_model_type(results, metric, normalization=None):
    x, y, generator = [], [], []
    for outcome in results:
        
        x.append(outcome['algo'])
        val = float(outcome[metric])
        if 'generator' in outcome:
            generator.append(outcome['generator'])
        else:
            generator.append('Real Networks')
            if normalization:
                val /= normalization
        y.append(val)
            
    return pd.DataFrame.from_dict({'Algorithm': x, metric: y, 'generator': generator})
```

```python
eval_SID_avg = average_performance(results, 'SID')
eval_SHD_avg = average_performance(results, 'SHD')
eval_ID_avg = average_performance(results, 'ID')
```

```python
eval_SID_generator = average_performance_model_type(full_results, 'SID', normalization=10)
```

```python
eval_SHD_generator = average_performance_model_type(full_results, 'SHD', normalization=5)
```

```python
eval_ID_generator = average_performance_model_type(full_results, 'ID')
```

```python
def plot_results(results_generator, results_avg, metric, fname=None):
    plt.close()
    sns.set_context('notebook')
    fig, axes = plt.subplots(1,2, figsize=(18,3), gridspec_kw={'width_ratios': [3, 1]}, sharey=True)
    
    ax = sns.barplot(x='generator', y=metric, ci=68, hue='Algorithm', palette=palette, ax=axes[0], data=results_generator)
    ax.legend_.remove()
    axes[0].set_xlabel('', fontsize=17)

    sns.barplot(x="Algorithm", y=metric, data=results_avg, ax=axes[1], ci=68, palette=palette)
    axes[1].set_xlabel('', fontsize=17)
    axes[1].set_ylabel('', fontsize=17)

    fig.tight_layout(pad=1.0)
    fig.legend(bbox_to_anchor=(0.5, 1.1), loc='upper center', ncol=8, fontsize='large')
    if fname:
        fig.savefig(fname, dpi=300)
    plt.show()
```
